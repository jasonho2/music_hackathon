```python
import pandas as pd
import numpy as np
import os

# path for reading
path = 'C:/Users/jason/Desktop/MillionSongSubset/data/A/A/A/'
data = pd.HDFStore(path+'TRAAAAW128F429D538.h5', mode='r')

# gets the columns names for the master dataframe
a = data.get('/analysis/songs')
b = data.get('/metadata/songs')
c = data.get('/musicbrainz/songs')

n = a.join(b, on=None, how='outer')
nn = n.join(c, on=None, how='outer')

columns = nn.columns

# in folder A, access two folder levels below and extract n .h5 files from each folder
# there is too much data (folders A-Z) so just stick with 1 folder and you can look at another later
# ...path/A/A-Z/A-Z/filename.h5
song_list = []

filepath = 'C:/Users/jason/Desktop/MillionSongSubset/data/A'

for a in os.listdir(filepath):
    for b in os.listdir(filepath+'/{}'.format(a)):
    
        for file in os.listdir(filepath+'/{}'.format(a)+'/{}'.format(b)):
            with open(filepath+'/{}'.format(a)+'/{}'.format(b)+'/{}'.format(file)):
                d = pd.HDFStore(filepath+'/{}'.format(a)+'/{}'.format(b)+'/{}'.format(file), mode='r')
                if not d:
                    break

            aa = d.get('/analysis/songs')
            bb = d.get('/metadata/songs')
            cc = d.get('/musicbrainz/songs')

            new = aa.join(bb, on=None, how='outer')
            newnew = new.join(cc, on=None, how='outer')
            
            song_list.append(newnew)
            
# to dataframe

song_df = pd.DataFrame(columns=columns)

for i in range(len(song_list)):
    temp_df = pd.DataFrame(song_list[i])
    song_df = song_df.append(temp_df)
    
song_df

# export

df = song_df
df.to_csv('C:/Users/jason/Desktop/data_science/songs3.csv')
```
